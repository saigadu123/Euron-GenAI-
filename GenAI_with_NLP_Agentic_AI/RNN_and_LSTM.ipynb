```python
input = [1,2,3,4]
output = [5,10,15,20]
```

```python
w = 0.0
b = 0.0
lr = 0.01
```

### Perceptron

```python
for epoch in range(100):
    total_loss = 0
    for x,y in zip(input,output):
        y_pred = w*x+b
        error = y_pred-y
        loss = error**2
        total_loss = total_loss + loss

        dw = 2*error*x
        db = 2*error

        w = w - lr*dw
        b = b - lr*db

    if epoch % 10 ==0:
        print(f"epoch: {epoch} | Loss: {total_loss} | w : {w} | b: {b}")

test_data = 10
prediction = w*test_data + b
print(prediction)
```

### Neural Network with hidden layer

```python
input = [1,2,3,4,5]
output = [2,4,6,8,10]
```

```python
w1,b1 = 0.1,0.0
w2,b2 = -0.1,0.0
w3,w4 = 0.1,0.1
b3 = 0.0
lr = 0.01
```

```python
def relu(input):
    return max(0,input)
```

```python
def relu_derivative(x):
    return 1 if x>0 else 0
```

![Neural Network](NN-euron.png)

```python
for epoch in range(100):
    total_loss = 0

    for x,y in zip(input,output):
        z1 = w1*x+b1
        h1 = relu(z1)

        z2 = w2*x+b2
        h2 = relu(z2)

        y_pred = w3*h1 + w4*h2 + b3
        error = y_pred-y
        loss = error*2
        total_loss = total_loss+loss

        dl_dy = 2*error

        dw3 = h1 * dl_dy
        dw4 = h2 * dl_dy
        db3 = dl_dy

        dz1 = dl_dy* w3 * relu_derivative(z1)
        dz2 =  dl_dy *w4 * relu_derivative(z2)

        dw1 = dz1 * x
        db1 = dz1

        dw2 = dz2 * x
        db2 = dz2 

        w1 = w1 - lr*dw1
        w2 = w2 - lr*dw2
        w3 = w3 - lr*dw3
        w4 = w4 - lr*dw4

        b1 = b1 - lr*db1
        b2 = b2 - lr*db2
        b3 = b3 - lr*db3

    
    if epoch%10 ==0:
        print(f"epoch: {epoch} | Loss: {total_loss} | ")


def prediction(x):
    z1 = w1*x + b1
    h1 = relu(z1)

    z2 = w2*x + b2
    h2 = relu(z2)

    y_pred = h1*w3+ h2*w4 + b3
    return y_pred

prediction(9)
```

```python
import tensorflow as tf
import numpy as np
```

```python
x = np.array(input,dtype=float)
y = np.array(output,dtype=float)
```

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(30,activation="relu",input_shape=[1]),
                    tf.keras.layers.Dense(1)])

```

```python
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),loss="mean_squared_error")
```

```python
model.fit(x,y,epochs=300,verbose=1)
model.predict(np.array([9]))
```
